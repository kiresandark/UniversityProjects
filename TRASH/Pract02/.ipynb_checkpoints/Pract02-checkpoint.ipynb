```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
tall = pd.read_csv("data.csv")
tall.head(8)

tall = tall.drop(['Name','Ticket','Cabin'], axis=1)
tall.head(8)


tall_aggr1=pd.pivot_table(tall,values='PassengerId', index='Pclass', columns='Survived', aggfunc='count')
tall_aggr1=tall.pivot_table(values='PassengerId', index='Pclass',
columns='Survived', aggfunc='count')
tall_aggr1=tall.pivot_table('PassengerId', 'Pclass', 'Survived', 'count')

tall_aggr1.plot(kind='bar', stacked=True)
plt.show()

tall_aggr1.plot(kind='bar', stacked=True, figsize=(16,10), fontsize =18, rot=0)
plt.show()

tall_aggr1.plot(kind='bar', stacked=True, figsize=(16,10), fontsize =18, rot=0)
plt.legend(labels=['Спаслось','Утонуло'], fontsize=18)
plt.xlabel("Cоциальный статус пассажира", fontsize=18)
plt.ylabel("Число пассажиров", fontsize=18)
plt.show()

tall_aggr1.plot(kind='bar', stacked=True, figsize=(16,10), fontsize =18, rot=0, colormap='winter')
plt.legend(labels=['Спаслось','Утонуло'], fontsize=18)
plt.xlabel("Cоциальный статус пассажира", fontsize=18)
plt.ylabel("Число пассажиров", fontsize=18)
plt.show()

plt.savefig("PClass.jpg", dpi=300)


writer = pd.ExcelWriter('AggrTitanic.xlsx', engine='xlsxwriter')
tall_aggr1.to_excel(writer, sheet_name='PClass', index=True, index_label = True)

tall_aggr2=tall.pivot_table('PassengerId', 'SibSp', 'Survived', 'count')
tall_aggr2=tall.pivot_table('PassengerId', 'SibSp', 'Survived', 'count', fill_value=0)
tall_aggr3=tall.pivot_table('PassengerId', 'Parch', 'Survived', 'count', fill_value=0)

fig, axes = plt.subplots(ncols=2)
tall_aggr2.plot(ax=axes[0], title='SibSp')
tall_aggr3.plot(ax=axes[1], title='Parch')
plt.show()

fig, axes = plt.subplots(ncols=2)
tall_aggr2.plot(kind='bar',ax=axes[0], title='SibSp')
tall_aggr3.plot(kind='bar',ax=axes[1], title='Parch')
plt.show()


tall_aggr2.to_excel(writer, sheet_name='SibSb', index=True, index_label = True)
tall_aggr3.to_excel(writer, sheet_name='Parch', index=True, index_label = True)
writer.save()

tall.to_csv('.Prak02.csv', index=False, sep=';',
float_format='%.5f', decimal=',' )
```
